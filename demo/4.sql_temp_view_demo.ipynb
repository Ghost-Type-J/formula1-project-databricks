
#### Access dataframes using SQL

#####Objectives
1. Create temporary views on dataframes
2. Access the view from a SQL cell
3. Access the view from a python cell

```
%run

"../includes/configuration/"
```

```
race_results_df = spark.read.parquet(f"{presentation_folder_path}/race_results")
```

```
race_results_df.createTempView("v_race_results")
```

```
%sql
SELECT COUNT(*)
FROM v_race_results
WHERE race_year = 2020
```

```
race_results_2019_df = spark.sql(
    "SELECT * FROM v_race_results WHERE race_year = 2019"
)
```

```
display(race_results_2019_df)
```
